```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the in-built Titanic dataset from seaborn
df = sns.load_dataset('titanic')

print("Dataset Loaded Successfully")
print(df.head())


print("\nDataset Summary:")
print(df.describe(include='all'))

# Select numerical columns
num_df = df.select_dtypes(include=np.number)

# Plot histograms for numerical variables
for col in num_df.columns:
    plt.figure()
    plt.hist(num_df[col].dropna())
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()

# Create a pairplot for numerical variables
sns.pairplot(num_df.dropna())
plt.show()

# Create a boxplot for multiple numerical variables
plt.figure(figsize=(10,6))
sns.boxplot(data=num_df)
plt.title("Boxplot for Multiple Numerical Variables")
plt.xticks(rotation=45)
plt.show()
```
