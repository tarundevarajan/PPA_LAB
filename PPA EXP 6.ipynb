```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder

# Define an inbuilt dataset as a list of transactions
transactions = [
    ['milk', 'bread', 'butter'],
    ['milk', 'sugar'],
    ['bread', 'butter'],
    ['milk', 'bread', 'sugar', 'eggs'],
    ['bread', 'milk', 'eggs']
]

print("Using an inbuilt sample dataset for demonstration.")
print(f"Total transactions loaded: {len(transactions)}")
if len(transactions) > 0:
    print("First 5 transactions:")
    for i in range(min(5, len(transactions))):
        print(f"  {transactions[i]}")
else:
    print("No transactions were loaded. The dataset might be empty or malformed.")

# One-hot encoding
te = TransactionEncoder()
te_data = te.fit(transactions).transform(transactions)
df_encoded = pd.DataFrame(te_data, columns=te.columns_)

# Apply Apriori
# Only proceed if df_encoded is not empty
if not df_encoded.empty:
    # Adjusted min_support for the smaller inbuilt dataset
    frequent_items = apriori(df_encoded, min_support=0.4, use_colnames=True)

    # Generate Association Rules
    # Only proceed if frequent_items is not empty
    if not frequent_items.empty:
        rules = association_rules(frequent_items, metric="confidence", min_threshold=0.3)

        # Display Results
        print("\nFrequent Itemsets:\n", frequent_items.head())
        print("\nAssociation Rules:\n", rules[['antecedents','consequents','support','confidence','lift']].head())
    else:
        print("\nNo frequent itemsets found with the given min_support. Try lowering min_support.")
else:
    print("\nEncoded DataFrame is empty. Cannot apply Apriori algorithm.")
```
